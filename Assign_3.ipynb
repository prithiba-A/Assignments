Q1. Create a file that contains 1000 lines of random strings.

```python
import random
import string

# Function to generate a random string
def generate_random_string(length):
    letters = string.ascii_letters
    return ''.join(random.choice(letters) for _ in range(length))

# Number of lines and string length
num_lines = 1000
string_length = 10

# Generate random strings and write them to a file
with open('random_strings.txt', 'w') as file:
    for _ in range(num_lines):
        random_string = generate_random_string(string_length)
        file.write(random_string + '\n')

print("File 'random_strings.txt' with 1000 random strings has been created.")
```

Q6. Convert all the files of Q4 into upper case parallel using multi-threading.

```python
import concurrent.futures

# Function to convert the content of a file to uppercase
def convert_file_to_uppercase(file_name):
    with open(file_name, 'r+') as file:
        content = file.read()
        file.seek(0)
        file.write(content.upper())
        file.truncate()

# Convert files to uppercase in parallel using multi-threading
with concurrent.futures.ThreadPoolExecutor() as executor:
    futures = []
    for i, size in enumerate(file_sizes):
        file_number = i + 1
        file_name = f"random_strings_{file_number}GB.txt"
        future = executor.submit(convert_file_to_uppercase, file_name)
        futures.append(future)

    # Wait for all tasks to complete
    concurrent.futures.wait(futures)

# Print confirmation messages
for i, size in enumerate(file_sizes):
    file_number = i + 1
    file_name = f"random_strings_{file_number}GB.txt"
    print(f"File '{file_name}' has been converted to uppercase in parallel.")
```

Q10. Create an automated pipeline using multi-threading for:
“Automatic Download of 100 Videos from YouTube” → “Convert it to Audio”

```python
import threading
import pytube
from pydub import AudioSegment

# Number of videos to download
num_videos = 100

# YouTube video URLs
video_urls = [
    "https://www.youtube.com/watch?v=VIDEO_ID_1",
    "https://www.youtube.com/watch?v=VIDEO_ID_2",
    # Add more YouTube video URLs here...
]

# Directory to save downloaded videos and audio
download_directory = "youtube_downloads"

# Semaphore to limit the number of concurrent downloads
download_semaphore = threading.Semaphore(5)

# Download video from YouTube
def download_video(url):
    try:
        # Acquire the download semaphore
        download_semaphore.acquire()

        # Create a pytube YouTube object
        yt = pytube.YouTube(url)

        # Get the video stream with the highest resolution
        video_stream = yt.streams.get_highest_resolution()

        # Download the video
        video_stream.download(output_path=download_directory)
        print(f"Video downloaded: {url}")

    finally:
        # Release the download semaphore
        download_semaphore.release()

# Convert video to audio
def convert_to_audio(video_path):
    audio_path = video_path[:-4] + ".mp3"  # Change file extension to .mp3

    # Load the video file
    video = AudioSegment.from_file(video_path, format="mp4")

    # Extract the audio from the video
    audio = video.set_channels(1)  # Convert to mono audio
    audio.export(audio_path, format="mp3")

    print(f"Video converted to audio: {video_path}")

# Create threads for downloading and converting
threads = []
for i in range(num_videos):
    video_url = video_urls[i % len(video_urls)]
    video_filename = f"video_{i+1}.mp4"
    video_path = f"{download_directory}/{video_filename}"

    # Download thread
    download_thread = threading.Thread(target=download_video, args=(video_url,))
    threads.append(download_thread)
    download_thread.start()

    # Convert to audio thread
    convert_thread = threading.Thread(target=convert_to_audio, args=(video_path,))
    threads.append(convert_thread)
    convert_thread.start()

# Wait for all threads to complete
for thread in threads:
    thread.join()

print("All videos downloaded and converted to audio successfully.")
```

Q11. Create an automated pipeline using multi-threading for: “Automatic Download of 500 images of Dog from
GoogleImages” → “Rescale it to 50%”

```python
import os
import threading
from google_images_download import google_images_download
from PIL import Image

# Number of images to download
num_images = 500

# Keyword for image search
keyword = "Dog"

# Directory to save downloaded and resized images
download_directory = "dog_images"

# Semaphore to limit the number of concurrent downloads
download_semaphore = threading.Semaphore(10)

# Download images from Google Images
def download_image(keyword):
    try:
        # Acquire the download semaphore
        download_semaphore.acquire()

        # Set the download arguments
        download_arguments = {
            "keywords": keyword,
            "limit": num_images,
            "output_directory": download_directory,
            "chromedriver": "chromedriver"  # Path to the chromedriver executable
        }

        # Create a google_images_download object
        response = google_images_download.googleimagesdownload()

        # Download the images
        response.download(download_arguments)

        print(f"Images downloaded for keyword: {keyword}")

    finally:
        # Release the download semaphore
        download_semaphore.release()

# Rescale images to 50%
def rescale_image(image_path):
    image = Image.open(image_path)

    # Calculate the new width and height
    width, height = image.size
    new_width = int(width * 0.5)
    new_height = int(height * 0.5)

    # Resize the image
    resized_image = image.resize((new_width, new_height), Image.ANTIALIAS)

    # Save the resized image, overwriting the original file
    resized_image.save(image_path)

    print(f"Image rescaled: {image_path}")

# Create threads for downloading and rescaling
threads = []
for i in range(num_images):
    # Download thread
    download_thread = threading.Thread(target=download_image, args=(keyword,))
    threads.append(download_thread)
    download_thread.start()

    # Rescale thread
    image_filename = f"{keyword}_{i+1}.jpg"
    image_path = os.path.join(download_directory, image_filename)
    rescale_thread = threading.Thread(target=rescale_image, args=(image_path,))
    threads.append(rescale_thread)
    rescale_thread.start()

# Wait for all threads to complete
for thread in threads:
    thread.join()

print("All images downloaded and rescaled successfully.")
```

Q13. Create a random dataset of 500 rows and 10 columns.
Columns 1 to 4 are defined between [-10, 10];
Columns 5 to 8 are defined between [10, 20];
Columns 9 to 10 are defined between [-100, 100].
Apply following clustering algorithms; determine the optimal number of clusters and plot distance metric
graph using each algorithm.
(i) K-Mean clustering
(ii) Hierarchical clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage

# Set random seed for reproducibility
np.random.seed(123)

# Create a random dataset of 500 rows and 10 columns
dataset = pd.DataFrame(np.random.uniform(low=-10, high=10, size=(500, 4)),
                       columns=['Column1', 'Column2', 'Column3', 'Column4'])
dataset[['Column5', 'Column6', 'Column7', 'Column8']] = np.random.uniform(low=10, high=20, size=(500, 4))
dataset[['Column9', 'Column10']] = np.random.uniform(low=-100, high=100, size=(500, 2))

# (i) K-Means clustering
distortions = []
silhouette_scores = []
max_clusters = 10

for k in range(2, max_clusters+1):
    kmeans = KMeans(n_clusters=k, random_state=123)
    kmeans.fit(dataset)
    distortions.append(kmeans.inertia_)
    silhouette_scores.append(kmeans.score(dataset))

# Plotting the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(2, max_clusters+1), distortions, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion')
plt.title('K-Means Clustering: Elbow Curve')
plt.show()

# Plotting the Silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, max_clusters+1), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('K-Means Clustering: Silhouette Scores')
plt.show()

# (ii) Hierarchical clustering
linkage_matrix = linkage(dataset, method='ward')

# Plotting the Dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix)
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering: Dendrogram')
plt.show()
```

Q15. Create a random dataset of 500 rows and 5 columns:
All the values are defined between [5,10].
Perform the following operations:
(i) Perform t-Test on each column.
(ii) Perform Wilcoxon Signed Rank Test on each column.
(iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp, wilcoxon, ttest_ind, ranksums

# Set random seed for reproducibility
np.random.seed(123)

# Create a random dataset of 500 rows and 5 columns
dataset = pd.DataFrame(np.random.uniform(low=5, high=10, size=(500, 5)),
                       columns=['Column1', 'Column2', 'Column3', 'Column4', 'Column5'])

# (i) Perform t-Test on each column
ttest_results = {}
for column in dataset.columns:
    t_statistic, p_value = ttest_1samp(dataset[column], 7.5)  # Assuming the population mean is 7.5
    ttest_results[column] = {'t_statistic': t_statistic, 'p_value': p_value}

# (ii) Perform Wilcoxon Signed Rank Test on each column
wilcoxon_results = {}
for column in dataset.columns:
    statistic, p_value = wilcoxon(dataset[column] - 7.5)  # Assuming the population median is 7.5
    wilcoxon_results[column] = {'statistic': statistic, 'p_value': p_value}

# (iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4
column3 = dataset['Column3']
column4 = dataset['Column4']
ttest_ind_result = ttest_ind(column3, column4)
ranksums_result = ranksums(column3, column4)

# Print the results
print('(i) t-Test Results:')
for column, result in ttest_results.items():
    print(f'{column}: t-statistic = {result["t_statistic"]:.4f}, p-value = {result["p_value"]:.4f}')

print('\n(ii) Wilcoxon Signed Rank Test Results:')
for column, result in wilcoxon_results.items():
    print(f'{column}: statistic = {result["statistic"]:.4f}, p-value = {result["p_value"]:.4f}')

print('\n(iii) Two Sample t-Test and Wilcoxon Rank Sum Test Results:')
print(f'Two Sample t-Test: t-statistic = {ttest_ind_result.statistic:.4f}, p-value = {ttest_ind_result.pvalue:.4f}')
print(f'Wilcoxon Rank Sum Test: statistic = {ranksums_result.statistic:.4f}, p-value = {ranksums_result.pvalue:.4f}')
```
