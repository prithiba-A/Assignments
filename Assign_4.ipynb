**Importing the required libraries for EDA**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pl
```

**Loading the dataset**

```python
data = pd.read_csv('/content/Wine (3).zip')
data
```

```python
data.head(5)
```

```python
data.tail(10)
```

```python
data.shape
```

**Checking the Data Types**

```python
data.dtypes
```

```python
data.info
```

```python
pd.unique(data['fixed acidity'])
```

**Dropping the columns**

```python
data = data.drop(['chlorides'], axis=1)
data.head(5)
```

**Renaming the Columns**

```python
data = data.rename(columns={"fixed acidity": "fixed acid", "volatile acidity": "volatile acid", "free sulfur dioxide": "free sulfur", "total sulfur dioxide": "total sulfur" })
data.head(5)
```

```python
data.shape
```

**Identify the Duplicate Rows**

```python
duplicate_rows_data = data[data.duplicated()]
print("number of duplicate rows: ", duplicate_rows_data.shape)
```

```python
data.count()
```

```python
data = data.drop_duplicates()
data.head()
```

```python
data.count()
```

**Identify the NULL Columns**

```python
data.isnull()
```

```python
print(data.isnull().sum())
```

**Dropping the missing values.**

```python
data = data.dropna()
data.count()
```

**BOX PLOTS**

```python
sns.boxplot(x=data['density'])
```

```python
sns.boxplot(x=data['pH'])
```

```python
sns.boxplot(x=data['alcohol'])
```

**Plot using Matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
data.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("fixed acidity	volatile vs total sulfur")
plt.ylabel('total sulfur')
plt.xlabel('fixed acidity	volatile')
```

**HIST PLOT**

```python
plt.figure(figsize=(10,6))
sns.histplot(data['pH'])
```

**SCATTER PLOT**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=data['residual sugar'], y=data['alcohol'])
```

**LINE PLOT**

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=data['density'],y=data['pH'])
```

**HEAT MAP**

```python
plt.figure(figsize=(10,5))
c= data.corr()
sns.heatmap(c,annot=True)
```
